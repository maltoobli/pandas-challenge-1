## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df['line_subtotal'] = df['unit_price'] * df['qty']
df[["unit_price", "qty", "line_subtotal"]].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

#first we have to calculate the total weight for the shippment:
df["total_weight"] = df["unit_weight"] * df["qty"]

#I used Lambda to calculate the price accourding to the weight
df['Shipping_Price'] = df["total_weight"].apply(lambda x: 7 * x if x > 50 else 10 * x)

#this one just to show the columns required as sample provided :
df[["unit_price","unit_weight","qty","total_weight","Shipping_Price"]].head(3)

```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%


#Here to find the total Price (line Price)
df["line_price"] = ((df["line_subtotal"] + df["Shipping_Price"]) * 1.0925).round(2)

#Here to show only the columns provided in sample answer- always can use df.head() to show all the df:
df[["line_subtotal", "Shipping_Price", "line_price"]].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df["line_cost"] = ((df["unit_cost"] * df["qty"]) + df["Shipping_Price"]).round(2)

df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price

df["line_profit"] = (df["line_price"] - df["line_cost"]).round(2)

#Here to show the top 3 rows as the sample provided:
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

top_five_clients = [33615, 66037, 46820, 38378, 24741]

for client_id in top_five_clients:
     ammount = df.loc[(df["client_id"]== client_id) , "line_price"].sum()
     print(f" {client_id} : ${ammount :.2f}")   
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

#List of clients and columns needed
top_five_clients = [33615, 66037, 46820, 38378, 24741]
columns_needed = ["qty","Shipping_Price", "line_price", "line_cost", "line_profit"]

summaries = {"client_id" : top_five_clients }

def summarize_clients(client_id,col_name):
    return df.loc[df["client_id"]== client_id , col_name].sum().round(2)
    
for col in columns_needed:
    summaries[col]= [summarize_clients(client_id,col) for client_id in top_five_clients]

    
summary_df = pd.DataFrame(summaries)
    
    
summary_df = summary_df.sort_values(by="line_profit", ascending=False)
    
summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

#rename the columns:
final_summary =summary_df.rename(columns = { "client_id" :" Client ID",
                                 "qty": "Units",
                                 "Shipping_Price" : "Shipping",
                                 "line_price" : "Total Revenue",
                                 "line_cost" : "Total Cost",
                                 "line_profit" : "Total Profit"
                                 })
#creat function to convert to million and format:
columns_to_convert_m = ["Shipping" , "Total Revenue","Total Cost", "Total Profit"]

def convert_mil(ammount):
    return "${:,.2f} M".format(ammount /1000000)

for col in columns_to_convert_m:
    final_summary[col] = final_summary[col].apply(convert_mil)
    
    
final_summary  
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
final_summary = final_summary.sort_values(by="Total Profit" , ascending=False)
final_summary
```
